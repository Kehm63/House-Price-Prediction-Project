```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the necessary dataset
data= pd.read_csv('house_data.csv')
```

```python
data.head()
```

```python
data.describe()
```

## Relationship between price and square foot

```python
plt.scatter(y=data['price'],x=data['sqft_living'])
plt.title('relationship between price and square foot')
```

## Define input (x) and output (y) data

```python
x= data.drop(['price','date','id'], axis=1)
y= data['price']
```

## Using a linear regression model to train the dataset

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

## split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

## Train the dataset

```python
model.fit(x_train,y_train)
```

## Score the model to test accuracy.

```python
model.score(x_test,y_test)
```
